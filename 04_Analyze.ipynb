## Importing libraries and loading data
*The data is the one that we clean during the Process phase.*

```python
#Importing libraries
import numpy as np
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the data
df = pd.read_csv("/content/drive/MyDrive/Dataset/cyclistic_clean.csv")
df
```

## Descriptive Analysis

```python
df.describe()
```

```python
#Percent of user type
percent_usertype = (df.groupby('user_type').size()/len(df) * 100).round(0)
percent_usertype
```

*64% of the total Cyclistic users are annual members while 36% are casual riders.*

```python
#Bike type preference per user type
user_bike= df.groupby(['user_type','bike_type'])['bike_type'].count()
user_bike
```

*Both annual members and casual riders prefer classic bikes. Only casual riders use docked bikes.*

```python
avg_usertype = (df.groupby('user_type')[['distance_travelled_km', 'ride_duration_min']].mean()).round(2)
avg_usertype
```

*Generally, casual riders have the longest average ride duration (23 minutes) compared with annual members (12 minutes). Both annual members and casual riders have almost the same average distance travelled.*

```python
avg_userbike = (df.groupby(['user_type', 'bike_type'])[['distance_travelled_km', 'ride_duration_min']].mean()).round(2)
avg_userbike
```

*Docked bikes have the longest average ride duration which only casual riders use. Classic bikes have the longest average ride duration for annual members.*

```python
#Number of users per day
count_day= df.groupby(['day_type','day_of_week'])['user_type'].count()
count_day
```

*Most trips are recorded on Saturday.*

```python
#Number of users per month
count_season= df.groupby(['season','month'])['user_type'].count()
count_season
```

*There are more trips during summer and at least during winter.*

```python
#User Type distibution per Route type
count_route= df.groupby(['route_type','user_type'])['user_type'].count()
count_route
```

*Annual members prefer one-way trip while casual riders prefer round trip.*
